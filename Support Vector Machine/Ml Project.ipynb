```python
from sklearn.preprocessing import StandardScaler
std = StandardScaler() # it is used to scale in unit variance
```

```python
x_train_std = std.fit_transform(x_train)
x_test_std = std.transform(x_test)
```

```python
from sklearn.svm import svc
```

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score #Importing SVC algorithm
```

```python
sv = SVC()
```

```python
sv.fit(x_train_std , y_train)
```

```python
y_pred = sv.predict(x_test_std) # This code will predict the outcomes with test data
```

```python
ac_sv = accuracy_score(y_test , y_pred) # To get the accuracy of model
```
