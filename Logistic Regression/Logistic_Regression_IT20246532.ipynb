```python
#define the formula to perform classification based on input features
formula = 'stroke ~ id + gender + age + hypertension + heart_disease + ever_married + work_type + Residence_type + avg_glucose_level + bmi + smoking_status'
```

```python
#define the logistic regression model
model = logit(formula = formula, data = train_data).fit()
```

```python
#summary of the model
model.summary()
```

```python
#display the model parameter
print(np.exp(model.params))
```

```python
#import classification metrics
from sklearn.metrics import classification_report, accuracy_score 
```

```python
#predict the logistic regression classifier model
prediction = model.predict(exog = test_data)
```

```python
#convert the predicted value 
cutoff = 0.7 
y_prediction = np.where(prediction > cutoff, 1, 0)
```

```python
#original lable
y_actual = test_data["stroke"]
```

```python
accuracy = accuracy_score(y_actual, y_prediction)
```

```python
print('Accuracy: %.2f' % accuracy + "%")
```
